```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


#simple example of logistic map 
def Logistic_Map(prms):
    alpha = prms["alpha"]
    steps = prms["steps"]
    ic = prms["ic"]
    y = np.zeros(steps)
    y[0] = ic
    
    for t in range(steps - 1):
        y[t + 1] = alpha * y[t] * (1 - y[t])
        
    return y

def plot_logistic_maps(alpha_values, steps, ic):
    sns.set_style("whitegrid")
    
    # Calculate number of rows and columns for subplots
    n = len(alpha_values)
    rows = int(np.ceil(n / 2))
    cols = 2

    fig, axes = plt.subplots(rows, cols, figsize=(20, 8 * rows))
    fig.suptitle("Logistic Map for Different α Values", fontsize=16)

    for i, alpha in enumerate(alpha_values):
        prms = {"alpha": alpha, "steps": steps, "ic": ic}
        y = Logistic_Map(prms)
        
        row = i // 2
        col = i % 2
        ax = axes[row, col] if rows > 1 else axes[col]
        
        sns.lineplot(x=range(steps), y=y, ax=ax)
        ax.set_title(f"α = {alpha}")
        ax.set_xlabel("Time Steps")
        ax.set_ylabel("Population")
        ax.set_ylim(0, 1)

    # Remove any unused subplots
    for i in range(n, rows * cols):
        row = i // 2
        col = i % 2
        fig.delaxes(axes[row, col] if rows > 1 else axes[col])

    plt.tight_layout()
    plt.show()

def main():
    #you can changhe these values and see what happens to the equation.
    alpha_values = [0.8, 1.6, 3.5, 3.9]
    steps = 100
    initial_condition = 0.2

    plot_logistic_maps(alpha_values, steps, initial_condition)

if __name__ == "__main__":
    main()
    
```
